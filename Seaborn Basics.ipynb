```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
# Distribution plot
sns.distplot(df['age'])
```

```python
# Histogram only, no kde line, custom # of bins
sns.distplot(df['age'], kde=False, bins=7)
```

```python
# Change size + color
plt.figure(figsize=(8,4))
sns.distplot(df['age'], kde=False, bins=7, color='red')
```

```python
df.head()
```

```python
# Count plot
# X axis = sex, data is our df
# Hue lets you integrate other cols
# Palette is how you change colors
sns.countplot(x='cp', data=df, hue='sex', palette='terrain')
```

```python
df.head()
```

```python
# Box plot
sns.boxplot(x='sex', y='age', data=df)
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```

```python
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', palette='Dark2', size='age')
```
